## 4. Menangani Nilai NULLL

**.isna().sum() untuk melihat apakah dataset memiliki nilai NULL**

```python
dataset_split.isna().sum()
```

**Untuk menangani nilai NULL pada dataset kita dapat menggunakan kelas SimpleImputer pada Scikit-Learn**

```python
from sklearn.impute import SimpleImputer
```

**Mengubah nilai NULL yang ada dengan Strategy most_frequent (Modus)**

```python
imputer1 = SimpleImputer(strategy="most_frequent")
dataset_split["Marital Status"] = imputer1.fit_transform(dataset_split[["Marital Status"]])
dataset_split.isna().sum()
```

**Mengubah nilai NULL yang ada dengan Strategy mean**

```python
imputer2 = SimpleImputer(strategy="mean")
dataset_split["Tumor Size"] = imputer2.fit_transform(dataset_split[["Tumor Size"]])
dataset_split.isna().sum()
```

**Mengubah nilai NULL yang ada dengan Strategy median**

```python
imputer3 = SimpleImputer(strategy="median")
dataset_split["Survival Months"] = imputer3.fit_transform(dataset_split[["Survival Months"]])
dataset_split.isna().sum()
```

**Nilai NULL Berhasil Ditangani**

```python
dataset_split.head(10)
```

# ENCODING

## 7. One Hot Encoding

```python
data_encoding = pd.read_csv("Breast_Cancer.csv")
```

**Menerapkan One-Hot Encoding menggunakan kelas OneHotEncoder dari Scikit-Learn**

Disini saya menggunakan atribut 'Marital Status'

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(sparse=False)
```

```python
Marital_Status_encoder = onehotencoder.fit_transform(data_encoding[['Marital Status']])
```

```python
Marital_Status = pd.DataFrame(Marital_Status_encoder)
data_encoding = data_encoding.join(Marital_Status)

data_encoding.head(10)
```

**Menerapkan One-Hot Encoding menggunakan get_dummies dari Pandas**

Disini saya menggunakan atribut 'Status'

```python
Status_encoder = pd.get_dummies(data_encoding[['Status']])

Status = pd.DataFrame(Status_encoder)
data_encoding = data_encoding.join(Status)

data_encoding.head(10)
```
