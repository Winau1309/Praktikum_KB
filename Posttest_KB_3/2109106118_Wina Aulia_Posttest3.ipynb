**Import library matplotlib, pandas dan seaborn**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Breast_Cancer.csv')
df_cancer = pd.read_csv('Breast_Cancer.csv')
df_cancer.head()
```

# 1. Grafik Batang

Kesimpulan dari dataset Breast Cancer menggunakan Grafik Batang rendah

Survival Months(Bulan Kelangsungan Hidup) berdasarkan Grade (Tingkat Kanker)
+ Grade 1 : Tingkat 1 atau tingkat rendah biasanya kanker tumbuh lebih lambat dan lebih kecil kemungkinannya untuk menyebar sehingga memiliki Survival Months paling tinggi
+ Grade 2 : Tingkat 2 atau tingkat menengah kanker tumbuh lebih cepat daripada kanker tingkat 1 sehingga memiliki Survival Months lebih rendah dari tingkat 1
+ Grade 3 : Tingkat 3 atau tingkat tinggi kanker yang tumbuh lebih cepat yang lebih mungkin menyebar sehingga memiliki Survival Months lebih rendah dari tingkat 1 dan 2
+ Grade 4 : Tingkat 4 berarti kanker telah menyebar (metastasis) di luar payudara dan ke bagian lain dari tubuh sehingga memiliki Survival Months paling rendah

```python
x = list(df["Grade"])
y = list(df["Survival Months"])

plt.figure(figsize=(5,5))
sns.barplot( x='Grade' , y='Survival Months', data=df)
plt.title('Survival Months berdasarkan Grade')
plt.xlabel("Grade")
plt.ylabel("Survival Months")
plt.show()
```

# 2. Grafik Pie

Kesimpulan dari dataset Breast Cancer menggunakan Grafik Pie: 

Status Alive/Dead Berdasarkan Usia Pasien
+ Dead(Meninggal): Data pasien kanker yang meninggal dunia memiliki persentase lebih rendah yaitu sebesar 15.3%
+ Alive(Hidup) : Data pasien kanker yang masih hidup memiliki persentase lebih besar yaitu sebesar 84.7%

```python
x = list(df["Age"])
y = list(df["Status"])

data_pie = df['Status'].value_counts().rename_axis('Status').reset_index(name='Age')

plt.figure(figsize=(6,6))
plt.pie(data_pie.Age, labels= data_pie.Status, startangle=100, autopct='%.1f%%', explode = (0.1, 0), shadow = True)
plt.title('Status Alive/Dead Berdasarkan Usia')
plt.show()
```

# 3. Grafik Box Plot

Kesimpulan dari dataset Breast Cancer menggunakan Grafik Boxplot:

Status Alive/Dead Berdasarkan Usia Pasien

Data pasien kanker yang meninggal dunia(Dead) memiliki keragaman data pengamatan atau tingkat penyebaran di usia yang lebih tua. Sedangkan, data pasien kanker yang hidup(Alive) memiliki keragaman data pengamatan atau tingkat penyebaran di usia yang lebih muda dari Data pasien kanker yang meninggal dunia.

```python
sns.boxplot(y="Age" , x="Status" , data=df)
plt.title('Status Alive/Dead Berdasarkan Usia')
plt.xlabel("Status")
plt.ylabel("Age")
plt.show ()
```

# 4. Grafik Histogram

Kesimpulan dari dataset Breast Cancer menggunakan Grafik Histogram:

Grafik histogram tersebut menggambarkan seberapa banyak orang yang memiliki ukuran tumor tertentu. Dalam grafik ini, kita dapat melihat seberapa sering ukuran tumor muncul pada sejumlah pasien. Misalnya, jika ada banyak batang tinggi di sekitar angka 300 pada sumbu x 1-10 , ini berarti bahwa banyak pasien memiliki tumor dengan ukuran 1-10 sekitar 300 pasien.

Dengan kata lain, grafik ini memberikan gambaran tentang seberapa umum ukuran tumor tertentu pada sekelompok pasien. Semakin tinggi batang pada angka tertentu, semakin banyak pasien yang memiliki ukuran tumor tersebut.unia

```python
plt.figure(figsize=(10,10))
plt.hist(df_cancer['Tumor Size'], bins=15, color='pink', edgecolor='purple')

plt.title('Histogram  Data Breast Cancer Berdasarkan Tumor Size')
plt.ylabel('Jumlah Pasien')
plt.xlabel('Ukuran Tumor')

plt.show()
```

# 5. Grafik Heat Map

Dari heatmap yang menunjukkan hubungan antar variabel numerik dalam dataset Breast Cancer, kita dapat melihat bahwa ada beberapa korelasi yang penting antara variabel-variabel tersebut.

Kesimpulannya, grafik heatmap membantu kita melihat bagaimana variabel-variabel dalam dataset berkaitan satu sama lain. Ini dapat membantu kita dalam menganalisis bagaimana perubahan dalam satu variabel dapat memengaruhi variabel lainnya.

```python
cor_matrix = df_cancer.corr()

sns.heatmap(cor_matrix, cmap='RdBu', annot=True, center=0)

plt.title('Heat Map Breast Cancer')
plt.xlabel("Variabel")
plt.ylabel("Variabel")

plt.show()
```
